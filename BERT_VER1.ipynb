## Prepare Dataset

```python
model_name = "distilbert-base-uncased"
#"bert-base-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
bert_model = DistilBertModel.from_pretrained(model_name)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('train.csv')
genre = data['genre'].tolist()
synop = data['synopsis'].tolist()
name = data['movie_name'].tolist()
nameSynop = []

for i in range(len(genre)):
  nameSynop.append(name[i].lower() + " : " + synop[i].lower())
```

```python
genreLabels = []
cnt = 0
GenreToLabel = dict()
for curGenre in genre:
  if curGenre not in GenreToLabel:
    GenreToLabel[curGenre] = cnt
    cnt = cnt + 1
  genreLabels.append(GenreToLabel[curGenre])
```

```python
datasetDict = dict()
#datasetDict = combineData(synop, genreLabels)
datasetDict = {'label': genreLabels, 'text': nameSynop}

dataset = Dataset.from_dict(datasetDict,split="train+test")
dataset = dataset.train_test_split(test_size=.3, shuffle=True)
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(model_name)

tokenized_datasets = dataset.map(tokenize_function, batched=True)

small_train_dataset = tokenized_datasets["train"].shuffle(seed=42)
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42)
```

```python
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=10)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="test_trainer",
    learning_rate = 1e-05,
    weight_decay = 0.0,
    num_train_epochs = 2.0,
    evaluation_strategy="epoch")
```

```python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```
